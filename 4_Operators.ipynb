#Ifelifelse

```python
i = 20
if i == 10:
  print("i is 10")
elif i == 15:
  print("i is 15")
elif i == 20:
  print("i is 20")
else:
  print("i is not present")
```

#Match Case

```python
num = 3
match num:
        # pattern 1
        case 1:
            print("One")
        # pattern 2
        case 2:
            print("Two")
        # pattern 3
        case 3:
            print("Three")
        # default pattern
        case _:
            print("Number not between 1 and 3")
```

#Comparison/Relational/Conditional Operators

```python
a = 9
b = 5
print(a > b)
print(a < b)
print(a == b)
print(a != b)
print(a >= b)
print(a <= b)
```
